# Data cleaning

```python
# restrict idp database to admin 2 boundaries, and identify latest idp count
df_idp = df_idp[df_idp["adminLevel"] == 2]

# latest date
latest_per_op = df_idp.groupby("operation")["reportingDate"].max()

df_idp = df_idp[df_idp.apply(lambda r: r["reportingDate"] == latest_per_op[r["operation"]], axis=1)].copy()
idp_admin2 = (
    df_idp
    .groupby(["admin1Name", "admin2Name"], as_index=False)
    .agg(idps=("numPresentIdpInd", "sum"),
         latest_date=("reportingDate", "max"))
)

idp_admin2.head()
```

```python
# cleaning admin1 and admin2 names for merging
idp_admin2["state_clean"] = idp_admin2["admin1Name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
idp_admin2["lga_clean"] = idp_admin2["admin2Name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
idp_admin2
```

```python
# cleaning names similarly in the boundaries df for merging
df_boundaries["state_clean"] = df_boundaries["adm1_name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
df_boundaries["lga_clean"] = df_boundaries["adm2_name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
df_boundaries
```

```python
# adding adm2_pcodes to the idp df
adm2_lookup = (
    df_boundaries[["lga_clean", "adm2_pcode"]]
    .drop_duplicates()
)
idp_admin2 = idp_admin2.merge(
    adm2_lookup,
    on="lga_clean",
    how="left"
)
idp_admin2
```

```python
df_facilities["type"].value_counts()
```

```python
# since I'm aiming to look at primary healthcare facilities, I only want to consider "primary", "primary health clinic", and "primary health center"
df_facilities["type"] = df_facilities["type"].replace({
    "Primary Health Clinic": "Primary",
    "Primary Health Center": "Primary"
})
df_facilities["type"].value_counts()
phcs = df_facilities[df_facilities["type"] == "Primary"]
phcs
```

```python
# clean adm1 and adm2 names again in the phcs df
phcs["state_clean"] = phcs["state_name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
phcs["lga_clean"] = phcs["lga_name"].astype(str).str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
```

```python
# adding adm2_pcodes to the phcs df
adm2_lookup = (
    df_boundaries[["lga_clean", "adm2_pcode"]]
    .drop_duplicates()
)
phcs = phcs.merge(
    adm2_lookup,
    on="lga_clean",
    how="left"
)
```

```python
# removing records that didn't get matched to a code
phcs = phcs.dropna(subset=["adm2_pcode"])
phcs
```

```python
# finding how many phcs there are per admin2 and adding to idp df
phc_numbers = (
    phcs["adm2_pcode"]
    .value_counts()
    .reset_index()
)

phc_numbers.columns = ["adm2_pcode", "phc_count"]
idp_admin2 = idp_admin2.merge(phc_numbers, on="adm2_pcode", how="left")
```

# Calculating PHC density for idp populations

I will calculate PHCs per 10,000 IDPs

```python
idp_admin2["phc_density"] = idp_admin2["phc_count"] / (idp_admin2["idps"] * 10000)
idp_admin2
```

```python
final_df = idp_admin2[["state_clean", "lga_clean", "phc_density"]].sort_values("phc_density")
final_df
```

```python
# making a map
df_geo["adm2_pcode"] = df_geo["adm2_pcode"].astype(str)

geo_merged = df_geo.merge(idp_admin2, on="adm2_pcode", how="left")
geo_merged.plot(column="phc_density", legend=True)
```
